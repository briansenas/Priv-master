```r
mtcars
```

```r
cmtcars <- mtcars %>%
  group_by(cyl) %>%
  summarise(across(c("mpg", "wt", "hp"), mean))

ggplot(mtcars, aes(x = mpg)) +
  facet_wrap(
    ~ cyl,
  ) +
  geom_histogram(
    aes(y = after_stat(density)),
    bins = 20,
    color = "black",
    fill = "white"
  ) +
  labs(
    title = "MPG per Cylinder",
    y = "Count of cars",
    x = "MPG"
  ) +
  geom_density(
    alpha = 0.2,
    fill = "blue",
  ) +
  geom_vline(
    data = cmtcars,
    aes(xintercept = mpg),
    linetype = "dashed", size = 1, color = "red"
  )


ggplot(mtcars, aes(x = hp)) +
  geom_histogram(
    aes(y = after_stat(density)),
    bins = 10,
    color = "black",
    fill = "white"
  ) +
  facet_wrap(
    ~ cyl,
  ) +
  labs(
    title = "HP per Cylinder",
    y = "Count of cars",
    x = "HP"
  ) +
  geom_density(
    alpha = 0.2,
    fill = "blue",
  ) +
  geom_vline(
    data = cmtcars,
    aes(xintercept = hp),
    linetype = "dashed", size = 1, color = "red"
  )

ggplot(mtcars, aes(x = wt)) +
  geom_histogram(
    aes(y = after_stat(density)),
    bins = 20,
    color = "black",
    fill = "white"
  ) +
  facet_wrap(
    ~ cyl,
  ) +
  labs(
    title = "Car Weight per Cylinder",
    y = "Count of cars",
    x = "Car Weight"
  ) +
  geom_density(
    alpha = 0.2,
    fill = "blue",
  ) +
  geom_vline(
    data = cmtcars,
    aes(xintercept = wt),
    linetype = "dashed", size = 1, color = "red"
  )
```
